# Importação Dos Pacotes

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importação Dos Dados

```python
url = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

data = pd.read_csv(url, compression = 'zip')
```

# Análise e Limpeza Dos Dados

```python
data.head()
```

```python
data.columns = data.columns.str.replace('-', '')
```

```python
data.rename(columns = {'droga':'composto'}, inplace=True)
```

```python
data.head()
```

# Desafio 1 - Ordenar o countplot dos top 5 compostos

```python
composto_count = data['composto'].value_counts()
top5_composto = composto_count.index[0:5]
data_top5_composto = data.query('composto in @top5_composto')
data_top5_composto
```

```python
order = composto_count[0:5].sort_values().index
order
```

# Desafio 2 - Melhorar a visualização

```python
#Criando figura e eixos
fig, ax = plt.subplots(figsize=(20,10)) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#criando uma palheta de 5 cores
sns.set_palette(sns.color_palette("Blues", 5))

#Criando o gráfico 
sns.countplot(x='composto', data=data_top5_composto, ax=ax, order=order)

#Plotando as legendas e mudando o tamanho das fontes
plt.title('5 Compostos Mais Utilizados', fontsize=20)
plt.xlabel('Compostos', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.xticks( fontsize=12)
plt.yticks(fontsize=12)
    
#Mostrando o gráfico
plt.show()
```

# Desafio 4 - Estudar as estatísticas retornadas no describe

```python
data.loc[:, 'g0':'g771'].describe()
```

# A função describe do pandas retorna para o usuário diversas medidas estatísticas que são muito úteis para o entendimento da distribuição das variáveis de um dataframe, quando se trata de valores numéricos essa função nos retorna o números de elementos como 'count', medidas de  tendência central como a média(mean) e a mediana(50%) que nos fornece informações sobre a distribuição, e principamente analisadas juntas podem dizer muito sobre a variável, já que um distribuição normal por exemplo possui essas duas medidas próximas e partir da diferença entre elas conseguimos extrair informações sobre a variável.

# Além disso essa função nos fornece medidas de dispersão como o desvio padrão(std), a principal medida utilizada para medir a dispersão de uma variável, quanto maior esse valor mais variados estão os valores dentro da série.

# Para completar ela nos dá o valor máximo(max) e valor mínimo(min) que são muito importantes para analisar eventuais outliers ou até eventuais soluções para o problema para filtrar pelas variáveis mais afetadas nesse caso, além do primeiro e terceiro quartis(25% e 75%) que estão ligados a concentração dos valores, analisados em conjunto com a média e a mediana da um entendimento onde realmente os valores estão concentrados e se há um tendência para um dos lados

# Desafio 5 - Refletir sobre a manipulação do tamanho das visualizações

```python
#Criando o gráfico 
sns.boxplot(x='g0', data=data, color='cornflowerblue')

#Plotando as legendas e mudando o tamanho das fontes
plt.title('Distribuição Da Variável g0', fontsize=10)
plt.ylabel('G-0', fontsize=10)
plt.xlabel('Unidade', fontsize=10)
plt.xticks( fontsize=10)
plt.yticks(fontsize=10)
    
#Mostrando o gráfico
plt.show()
```

# Ao criar o boxplot sem definir um tamanho o próprio seaborn escolhe o tamanho padrão e a partir do valor máximo e mínimo da variável ele plota a escala para conseguirmos visualizar todos os elementos, nesse caso isso pode gerar um problema de visualização aparentando um gráfico distorcido, já que os valores estão concentrados em um intervalo muito pequeno de valores ao redor do 0 e os valores máximo e mínimo são valores muito desproporcionais e discrepantes se comparados às medidas de tendência central como média e mediana.Para solucionar esse problema podemos aumentar o tamanho da figura onde o seaborn plota as informações, pois dessa forma os valores estarão numa escala mais apropriada para visualização.

```python
#Criando figura e eixos
fig, ax = plt.subplots(figsize=(20,10)) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#Criando o gráfico 
sns.boxplot(y='g0', data=data, ax=ax, color='cornflowerblue')

#Plotando as legendas e mudando o tamanho das fontes
plt.title('Distribuição Da Variável g0', fontsize=20)
plt.ylabel('Unidade', fontsize=15)
plt.xlabel('G-0', fontsize=15)
plt.xticks( fontsize=12)
plt.yticks(fontsize=12)
    
#Mostrando o gráfico
plt.show()
```

Aumentando o tamanho do gráfico vemos que a visulização ficou mais clara 

# Desafio 6 - Fazer outras análises com boxplot e histograma

```python
#Criando figura e eixos
fig, ax = plt.subplots(3,3,figsize=(20,10), sharex=True, sharey=True) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#Criando o gráfico 
sns.boxplot(y='g0', data=data, ax=ax[0,0], color='cornflowerblue')
sns.boxplot(y='g1', data=data, ax=ax[0,1], color='cornflowerblue')
sns.boxplot(y='g2', data=data, ax=ax[0,2], color='cornflowerblue')
sns.boxplot(y='g3', data=data, ax=ax[1,0], color='cornflowerblue')
sns.boxplot(y='g4', data=data, ax=ax[1,1], color='cornflowerblue')
sns.boxplot(y='g5', data=data, ax=ax[1,2], color='cornflowerblue')
sns.boxplot(y='g6', data=data, ax=ax[2,0], color='cornflowerblue')
sns.boxplot(y='g7', data=data, ax=ax[2,1], color='cornflowerblue')
sns.boxplot(y='g8', data=data, ax=ax[2,2], color='cornflowerblue')

#Plotando as legendas e mudando o tamanho das fontes
fig.suptitle('Boxplots Das 9 Primeiras Variáveis Com Prefixo c', fontsize=20)
    
#Mostrando o gráfico
plt.show()
```

Podemos perceber que a distribuição entre as variáveis com prefixo 'g' se assemelham muito na distribuição, com grandes concentrações perto do 0

```python
#criando uma palheta de 5 cores
sns.set_palette(sns.color_palette("gnuplot2"))

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#Criando o gráfico 
sns.catplot(y='g0', data=data, kind='box', col='tempo', row='dose')

#Plotando as legendas e mudando o tamanho das fontes
fig.suptitle('Boxplots Da Variável g0', fontsize=20)
    
#Mostrando o gráfico
plt.show()
```

Não é possível olhando só para o 'g0' notar uma relação dessa variável com dose e tempo 

```python
data_legend = data.loc[:, ['tratamento', 'g0']]
data_legend['tratamento'] = data_legend['tratamento'].str.replace('com_droga', 'Com Composto')
data_legend['tratamento'] =  data_legend['tratamento'].str.replace('com_controle', 'Com Controle')
data_legend
```

```python
#Criando figura e eixos
fig, ax = plt.subplots(figsize=(20,10)) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#criando uma palheta de 3 cores
sns.set_palette(sns.color_palette('prism', 2))

#Criando o gráfico 
sns.boxplot(x='tratamento', y='g0', data=data_legend, ax=ax)

#Plotando as legendas e mudando o tamanho das fontes
plt.title('Distribuição Da Variável g0 Pelo Tipo De Tratamento', fontsize=20)
plt.ylabel('Unidade', fontsize=15)
plt.xlabel('G-0', fontsize=15)
plt.xticks( fontsize=12)
plt.yticks(fontsize=12)
    
#Mostrando o gráfico
plt.show()
```

Notamos que por mais que a concentração do 'g0' continua perto do 0 variando o tratamento, percebemos que há muito mais outliers no tratamento com composto, isso pode ser apenas por ter muito mais experimentos com compostos do que com controle, ou pode indicar alguma relação

```python
data_legend_composto = data_legend.loc[data_legend['tratamento'] == 'Com Composto']
data_legend_controle = data_legend.loc[data_legend['tratamento'] == 'Com Controle']
```

```python
#Criando figura e eixos
fig, ax = plt.subplots(1, 2,figsize=(20,10), sharex=True, sharey=True) 

#Criando o gráfico
sns.distplot(data_legend_composto['g0'], ax=ax[0],color='r')
sns.distplot(data_legend_controle['g0'], ax=ax[1], color='y')


#Plotando as legendas e mudando o tamanho das fontes
fig.suptitle('Distribuição Das 9 Primeiras Variáveis Com Prefixo c', fontsize=20)
    
#Mostrando o gráfico
plt.show()
```

Observando agora o histograma da mesma variável 'g0' percebemos que realmente a diferença está no intervalo  de valores, quando submetido ao composto essa variável teve muito mais valores discrepantes, podendo indicar a ação desse composto
